```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
params = {'legend.fontsize':20,
         'figure.titlesize':25,
         'figure.figsize':(20,10),
         'axes.labelsize':15,
         'axes.titlesize':20,
         'xtick.labelsize':30,
         'ytick.labelsize':30
         }

plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style('white')

%matplotlib inline
```

## Scatterplot

```python
#import dataset
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")
midwest.head()
```

```python
# data preparations
categories = np.unique(midwest['category'])
print(categories)

# create colors for categories
colors = [plt.cm.tab10(i/float(len(categories)-1)) for i in range(len(categories))]

# draw plot 
plt.figure(figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')

# draw plot for each category
for i, category in enumerate(categories):
    plt.scatter('area', 'poptotal', data=midwest.loc[midwest.category==category, :], s=100, c=colors[i], label=str(category))
    
    
# decorating the plot
plt.title('Scatterplot of midwest Area vs Population', fontsize=25, pad=60)
plt.xlabel(xlabel='Area', fontsize=25, labelpad=50, ha='left', rotation=0)
plt.ylabel(ylabel='Population', fontsize=25, labelpad=50, ha='right', rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20, )
plt.show()
```
