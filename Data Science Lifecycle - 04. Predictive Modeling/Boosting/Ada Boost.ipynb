# Ada Boost 
 - The basic concept behind Adaboost is to set the weights of classifiers and training the data sample in each iteration such that it ensures the accurate predictions of unusual observations.

 - Any machine learning algorithm can be used as base classifier if it accepts weights on the training set.

## Ada v/s Random Forest

1. Ada boost Combines lots of weak learners in the form of Stumps; Stumps are the tree with only Root node and two leaves.
2. Some stumps have greater influence than other stumps in Combination of stumps; unlike in random forest each tree have same weightage of say to vote.
3. Each stump is made by taking into account the mistake made by previous stump; unlike in random forest each tree is made independently.
4. Only one feature is taken into account while making the the stumps.


## Steps in Ada Boost

1. Create 3 Stumps for **Chest Pain**, **Blocked Arteries** & **Patient Weight** respectively, Calculate Gini index for each and pick up the stump having lowest Gini index say **Patient Weight**, Also assign sample weights to records. 

### DataSet


 
2. Suppose stump made mistake at 4th row then weight of this row is upgraded and other rows is downgrded by using following formulae.
   1. Performance of Say = `(1/2)*(log((1-Total Error)/Total Error))`
   2. New Weight for Mistakenly placed records = `Sample Weight * e^(Performance of Say)` 
   3. New Weight for Correctly placed records = `Sample Weight * e^(-Performance of Say)`
   4. New Weights does not sum up to give summation as 1 hence they are normalised by dividing by sum of total weights and this normalised weights are replaced with original weights of data.


3. Now new dataset is created which has many records like record which was incorrectly classified previously.and sample weightages are restored to initial sample weights like (1/number of records).It influences the next stump such that it corrects the previous mistake.

### New DataSet



4. Step 1 to 3 are repeated for creating new stumps which Corrects the mistake of previous stump by taking new dataset with updated weightages. these steps are followed till we get greater accuracy and all the records get placed correctly.

5. Suppose their are now 5 stumps which used as forest to make prediction, now every stump makes predictions and based on their weightages or amount of say of Stumps their predictions are made.
Weightage of say of stump in final classification depends upon Total Error of stump 
<br>      `Amount of say = (1/2)*log((1-Total Error)/(Total Error))`
       1. If T.E. == 0.5 then Amount of say is 0
       2. If T.E. == 1 then Amount of say is Large negative value
       3. If T.E. == 0 then amount of say is Positive 
    
    Hence Amount of say in classification varies with logarithmic function of Total Error.

## Advantages of AdaBoost 

1. AdaBoost is easy to implement.

2. It iteratively corrects the mistakes of the weak classifier and improves accuracy by combining weak learners.

3. We can use many base classifiers with AdaBoost.

4. AdaBoost is not prone to overfitting.

## Disadvantages of AdaBoost 

1. AdaBoost is sensitive to noise data.

2. It is highly affected by outliers because it tries to fit each point perfectly.

3. AdaBoost is slower compared to XGBoost.

# 3. Ada Boost Model 

Parameters
-----------------
`base_estimator`, `n_estimators`,`loss` and `learning_rate`.

1. `base_estimator` - is the learning algorithm to use to train the weak models. This will almost always not needed to be changed because by far the most common learner to use with AdaBoost is a decision tree – this parameter’s default argument.

2. `n_estimators` - is the number of models to iteratively train.

3. `learning_rate` - is the contribution of each model to the weights and defaults to 1. Reducing the learning rate will mean the weights will be increased or decreased to a small degree, forcing the model train slower (but sometimes resulting in better performance scores).

4. `loss` - is exclusive to AdaBoostRegressor and sets the loss function to use when updating weights. This defaults to a linear loss function however can be changed to square or exponential.

## 1. Optimum value of `n_estimators` parameter 

```python
scores=[]

for each in range(1,50):
    ada = AdaBoostClassifier(n_estimators=each, random_state=42)
    ada.fit(X_train, y_train)
    scores.append(ada.score(X_test, y_test))
    

plt.figure(1, figsize=(10,5))
plt.plot(range(1,50), scores, color="black", linewidth=2)
plt.title("Optimum n_estimators parameters value")
plt.xlabel("N Estimators")
plt.ylabel("Score in Accuracy")
plt.grid(True)
plt.show()
```

## We can see that their is highest accuracy at `n_estimators=3` hence use this value.

## 3.2 Ada Boost classifier Model

```python
decision_tree = DecisionTreeClassifier()

ada = AdaBoostClassifier(base_estimator=decision_tree, n_estimators=3, random_state=42, learning_rate=0.01)

ada.fit(X_train, y_train)

y_pred = ada.predict(X_test)
```

## 3.3 Evaluation of Model

```python
ada.score(X_test, y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True)
plt.xlabel("Prediction")
plt.ylabel("True value")
```

# 4. Hyperparameters Tuning

```python
decision_tree = DecisionTreeClassifier(max_depth=2, random_state=42)

ada = AdaBoostClassifier(decision_tree)
                             
params = {"base_estimator__criterion":["gini","entropy"],
         "base_estimator__splitter":["best", "random"],
         "algorithm": ["SAMME", "SAMME.R"],
         "n_estimators":[10,100,200,250],
         "learning_rate":[0.05,0.5,1.5,2.5]
         }

ada_cv = GridSearchCV(ada, params, refit=True, scoring="accuracy", n_jobs=2, verbose=1)
```

```python
ada_cv.fit(X_train, y_train)
```

```python
ada_cv.best_params_
```

```python
y_pred = ada_cv.predict(X_test)
```

# 5. Evaluation of hyperparameters tunned model

```python
ada_cv.score(X_test, y_test)
```

```python
cv = confusion_matrix(y_test, y_pred)

sns.heatmap(cv, annot=True)
plt.xlabel("Predictions")
plt.ylabel("True Values")
plt.show()
```

# 6. Using Model for Prediction

```python
X = X_train_df.drop("target", axis=1)
X = X.columns.tolist()
X_input = []


for i in X:
    para = input("Input "+i)
    X_input.append(para)

# Converting to 2D List for processing
X_input = [X_input]

#Predicting using model
y = ada_cv.predict(X_input)

# Printing the results
result = "Positive"
if y == 0:
    result = "Negative"

print("The result of heart disease is "+ result)
```
