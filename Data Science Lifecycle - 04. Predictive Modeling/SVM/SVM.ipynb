```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.read_csv('UniversalBank.csv')
data.head()
```

```python
# Dropping unnecessar coluns
data.drop(['ID','ZIP Code'], axis=1, inplace=True)
data.head()
```

```python
# splitting data

X = data.copy().drop('Personal Loan', axis=1)
Y = data["Personal Loan"]

trainX, testX, trainY, testY = train_test_split(X,Y, test_size=0.2)

print(trainX.shape)
print(trainY.shape)
print(testX.shape)
print(testY.shape)
```

```python
# training the model 
svc = SVC()

# fit
svc.fit(trainX, trainY)

# predict
train_predictions = svc.predict(trainX)
test_predictions = svc.predict(testX)

# Train data accuracy 
accuracy_score_train_data = accuracy_score(trainY, train_predictions)
accuracy_score_test_data  = accuracy_score(testY, test_predictions)

print(f"Train Data Accuracy {accuracy_score_train_data}")
print(f"Test Data Accuracy {accuracy_score_test_data}")
```
